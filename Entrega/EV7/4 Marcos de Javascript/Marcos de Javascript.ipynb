```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python


# Configurar el estilo de Seaborn
sns.set_theme(style="darkgrid")

# Cargar el archivo Excel
data_excel = pd.read_excel('/content/graficos/multiTimeline-_4_.xlsx')

# Convertir la columna 'Mes' a datetime
data_excel['Mes'] = pd.to_datetime(data_excel['Mes'], format='%Y-%m')

# Crear una nueva columna 'Ao' extrada de la columna 'Mes'
data_excel['Ao'] = data_excel['Mes'].dt.year

# Agrupar los datos por ao y calcular la media para cada tecnologa
data_excel_grouped = data_excel.groupby('Ao').mean().reset_index()

# Crear el grfico de correlacin lineal
plt.figure(figsize=(15, 8))

# Regresin para jQuery
sns.regplot(
    x=data_excel_grouped['Ao'],
    y=data_excel_grouped['jquery: (Todo el mundo)'],
    label='jQuery',
    scatter_kws={'s': 50},  # Tamao de los puntos
    line_kws={'color': 'blue'},  # Color de la lnea de regresin
)

# Regresin para AngularJS
sns.regplot(
    x=data_excel_grouped['Ao'],
    y=data_excel_grouped['angularjs: (Todo el mundo)'],
    label='AngularJS',
    scatter_kws={'s': 50},
    line_kws={'color': 'green'},  # Color de la lnea de regresin
)

# Regresin para Angular
sns.regplot(
    x=data_excel_grouped['Ao'],
    y=data_excel_grouped['angular: (Todo el mundo)'],
    label='Angular',
    scatter_kws={'s': 50},
    line_kws={'color': 'red'},  # Color de la lnea de regresin
)

# Regresin para React
sns.regplot(
    x=data_excel_grouped['Ao'],
    y=data_excel_grouped['React: (Todo el mundo)'],
    label='React',
    scatter_kws={'s': 50},
    line_kws={'color': 'purple'},  # Color de la lnea de regresin
)

# Configuracin del grfico
plt.title('Correlacin Lineal: Popularidad de Tecnologas (2010-2024)', fontsize=14, pad=20)
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Popularidad de Bsquedas (Promedio)', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(data_excel_grouped['Ao'], rotation=45)
plt.tight_layout()

# Mostrar el grfico
plt.show()

# Calcular las correlaciones entre las tecnologas
correlation_jquery_angularjs = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['angularjs: (Todo el mundo)']
)
correlation_jquery_angular = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['angular: (Todo el mundo)']
)
correlation_jquery_react = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)
correlation_angularjs_angular = data_excel_grouped['angularjs: (Todo el mundo)'].corr(
    data_excel_grouped['angular: (Todo el mundo)']
)
correlation_angularjs_react = data_excel_grouped['angularjs: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)
correlation_angular_react = data_excel_grouped['angular: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)

# Imprimir las correlaciones
print(f"Correlacin entre jQuery y AngularJS: {correlation_jquery_angularjs:.2f}")
print(f"Correlacin entre jQuery y Angular: {correlation_jquery_angular:.2f}")
print(f"Correlacin entre jQuery y React: {correlation_jquery_react:.2f}")
print(f"Correlacin entre AngularJS y Angular: {correlation_angularjs_angular:.2f}")
print(f"Correlacin entre AngularJS y React: {correlation_angularjs_react:.2f}")
print(f"Correlacin entre Angular y React: {correlation_angular_react:.2f}")
```
