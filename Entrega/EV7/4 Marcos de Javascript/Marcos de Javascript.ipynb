```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python


# Configurar el estilo de Seaborn
sns.set_theme(style="darkgrid")

# Cargar el archivo Excel
data_excel = pd.read_excel('/content/graficos/multiTimeline-_4_.xlsx')

# Convertir la columna 'Mes' a datetime
data_excel['Mes'] = pd.to_datetime(data_excel['Mes'], format='%Y-%m')

# Crear una nueva columna 'Año' extraída de la columna 'Mes'
data_excel['Año'] = data_excel['Mes'].dt.year

# Agrupar los datos por año y calcular la media para cada tecnología
data_excel_grouped = data_excel.groupby('Año').mean().reset_index()

# Crear el gráfico de correlación lineal
plt.figure(figsize=(15, 8))

# Regresión para jQuery
sns.regplot(
    x=data_excel_grouped['Año'],
    y=data_excel_grouped['jquery: (Todo el mundo)'],
    label='jQuery',
    scatter_kws={'s': 50},  # Tamaño de los puntos
    line_kws={'color': 'blue'},  # Color de la línea de regresión
)

# Regresión para AngularJS
sns.regplot(
    x=data_excel_grouped['Año'],
    y=data_excel_grouped['angularjs: (Todo el mundo)'],
    label='AngularJS',
    scatter_kws={'s': 50},
    line_kws={'color': 'green'},  # Color de la línea de regresión
)

# Regresión para Angular
sns.regplot(
    x=data_excel_grouped['Año'],
    y=data_excel_grouped['angular: (Todo el mundo)'],
    label='Angular',
    scatter_kws={'s': 50},
    line_kws={'color': 'red'},  # Color de la línea de regresión
)

# Regresión para React
sns.regplot(
    x=data_excel_grouped['Año'],
    y=data_excel_grouped['React: (Todo el mundo)'],
    label='React',
    scatter_kws={'s': 50},
    line_kws={'color': 'purple'},  # Color de la línea de regresión
)

# Configuración del gráfico
plt.title('Correlación Lineal: Popularidad de Tecnologías (2010-2024)', fontsize=14, pad=20)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Popularidad de Búsquedas (Promedio)', fontsize=12)
plt.legend(fontsize=10)
plt.xticks(data_excel_grouped['Año'], rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()

# Calcular las correlaciones entre las tecnologías
correlation_jquery_angularjs = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['angularjs: (Todo el mundo)']
)
correlation_jquery_angular = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['angular: (Todo el mundo)']
)
correlation_jquery_react = data_excel_grouped['jquery: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)
correlation_angularjs_angular = data_excel_grouped['angularjs: (Todo el mundo)'].corr(
    data_excel_grouped['angular: (Todo el mundo)']
)
correlation_angularjs_react = data_excel_grouped['angularjs: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)
correlation_angular_react = data_excel_grouped['angular: (Todo el mundo)'].corr(
    data_excel_grouped['React: (Todo el mundo)']
)

# Imprimir las correlaciones
print(f"Correlación entre jQuery y AngularJS: {correlation_jquery_angularjs:.2f}")
print(f"Correlación entre jQuery y Angular: {correlation_jquery_angular:.2f}")
print(f"Correlación entre jQuery y React: {correlation_jquery_react:.2f}")
print(f"Correlación entre AngularJS y Angular: {correlation_angularjs_angular:.2f}")
print(f"Correlación entre AngularJS y React: {correlation_angularjs_react:.2f}")
print(f"Correlación entre Angular y React: {correlation_angular_react:.2f}")
```
