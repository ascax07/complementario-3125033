```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


# Cargar el archivo Excel
data_excel = pd.read_excel('/content/idiomas_populares/multiTimeline_5.xlsx')

# Convertir la columna 'Semana' a datetime
data_excel['Semana'] = pd.to_datetime(data_excel['Semana'])

# Seleccionar las columnas de los lenguajes de programación
languages = [
    'Java: (Colombia)',
    'Python: (Colombia)',
    'javascript: (Colombia)',
    'php: (Colombia)'
]

# Crear el gráfico de correlación (scatter plot matrix)
sns.set(style="ticks")
sns.pairplot(data_excel[languages], height=2.5, aspect=1.2,
             plot_kws={'alpha': 0.7, 's': 30, 'edgecolor': 'k'},
             diag_kws={'bins': 30})

# Configurar título general
plt.suptitle('Correlación entre Lenguajes de Programación en Colombia', y=1.02)

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()

# Calcular y mostrar la matriz de correlación
correlation_matrix = data_excel[languages].corr()
print("\nMatriz de correlación:")
print(correlation_matrix)
```
